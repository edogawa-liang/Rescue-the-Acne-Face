# 重命名照片與比對資料夾檔案
1. 將照片檔案重新命名
2. 將兩資料夾重複名稱的照片挑出，用於選出篩選後的inpainting結果，作為側臉無痘痘照訓練Pix2Pix

```python
import os   
import glob
import shutil
```

```python
acne_path = 'stable_diffusion_inpainting挑/trainA_acne'
inpaint_path = 'stable_diffusion_inpainting挑/trainA_inpainting'

path_choose = '挑好的'
if not os.path.exists(path_choose+'/trainA_inpainting') :
    os.makedirs(path_choose+'/trainA_inpainting') 
if not os.path.exists(path_choose+'/trainA_acne') :
    os.makedirs(path_choose+'/trainA_acne') 
```

```python
acne = glob.glob(acne_path + r'/*.jpg', recursive=True)
inp = glob.glob(inpaint_path + r'/*.jpg', recursive=True)
```

```python
# 取acne inp 的交集
acne_name = []
inp_name = []
for i in range(len(acne)):
    acne_name.append(acne[i].split('/')[-1])
for i in range(len(inp)):
    inp_name.append(inp[i].split('/')[-1])

# 取兩個list的交集
common = list(set(acne_name).intersection(set(inp_name)))
print(len(common))
    
```

```python
# 重命名文件
for old_name in acne_name:
    # 生成新文件名，假設需要將 'levle1_1_' 替換為 'levle1_'
    # new_name = old_name.replace('levle1_1_', 'levle1_')
    # new_name = old_name.replace('levle1_2_', 'levle2_')
    # new_name = old_name.replace('levle1_0_', 'levle0_')
    new_name = old_name.replace('levle', 'level')
    
    # 檢查原文件是否存在
    if os.path.exists(f'{acne_path}/{old_name}'):
        # 重命名文件
        os.rename(f'{acne_path}/{old_name}', f'{acne_path}/{new_name}')
        print(f"{old_name} 已重命名為 {new_name}")
    else:
        print(f"未找到文件：{old_name}")
```

```python
# 將重疊的檔案複製到挑好的資料夾
for i in range(len(common)):
    shutil.copy(acne_path+'/'+common[i], path_choose+'/trainA_acne')
    shutil.copy(inpaint_path+'/'+common[i], path_choose+'/trainA_inpainting')
    print(common[i])
```
